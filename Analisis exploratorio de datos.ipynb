### Llamar los datos

```python
#Llamamos el set de datos público 
zoo_data = 'https://archive.ics.uci.edu/ml/machine-learning-databases/zoo/zoo.data'

# los nombres de las columnas se encuentran en :  'https://archive.ics.uci.edu/ml/machine-learning-databases/zoo/zoo.names'
zoo_name= ['animal name',
           'hair',
           'feathers',
           'eggs',
           'milk',
           'airborne',
           'aquatic',
           'predator',
           'toothed',
           'backbone',
           'breathes',
           'venomous',
           'fins',
           'legs',
           'tail',
           'domestic',
           ' cat-size',
           'type']
```

```python
import pandas as pd
```

```python
pd.read_csv(zoo_data)
```

```python
df_zoo=pd.read_csv(zoo_data)
```

Definimos los nombres de las columnas.

```python
df_zoo.columns=zoo_name
```

```python
df_zoo.head()
```

Graficando
> La siguiente librería es muy útil ya que genera imagenes de vectores para informes que pueden ser modificados por diseñadores.

```python
import altair as alt
```

```python
#librerias de graficación más empleadas 
import plotly as plt 
import seaborn as sns
```

```python
plot_1 = alt.Chart(df_zoo).mark_line().encode(
    x='animal name',
    y= 'legs'
)
```

```python
plot_1
```

Otro tipo de gráfica
> Imagen de vectores para informes.

```python
plot_2 = alt.Chart(df_zoo).mark_bar().encode(
    x='animal name',
    y= 'legs'
)
```

```python
plot_2
```

# Otras librerias

```python
import plotly.express as px
```

```python
fig = px.pie(df_zoo, values='aquatic', names='animal name', title='Porcentaje de animales acuáticos') # gráfica de pie 
fig.show() 
```

```python
sns.displot(df_zoo)
```

```python
sns.displot(df_zoo, x="airborne", kind="kde") # distribuciones 
```
